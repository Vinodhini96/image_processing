## apply filter to the image
module - from PIL import ImageFilter

```python
#to convert image into greyscale
gr_im = im.convert("L")
gr_im.show()  
plt.imshow(gr_im)
plt.show()
```

```python
#to covert imgae into blur
blur_im = im.filter(ImageFilter.BLUR)

plt.imshow(blur_im)
plt.show()
```

```python
#to contour the image
contour_im = im.filter(ImageFilter.CONTOUR)
plt.imshow(contour_im)
plt.show()
```

```python
# filter : detail
detail_im = im.filter(ImageFilter.DETAIL)
plt.imshow(detail_im)
plt.show()
```

```python
#filter : edge_enhance
edge_enhance_im= im.filter(ImageFilter.EDGE_ENHANCE)

plt.imshow(edge_enhance_im)
plt.show()
```

```python
#filter :EDGE_ENHANCE_MORE
enhance_more_im = im.filter(ImageFilter.EDGE_ENHANCE_MORE)

plt.imshow(enhance_more_im)
plt.show()
```

```python
#filter : EMBOSS
emboss_im = im.filter(ImageFilter.EMBOSS)

plt.imshow(emboss_im)
plt.show()
```

```python
# to smoothen the image 
smooth_im = im.filter(ImageFilter.SMOOTH)

plt.imshow(smooth_im)
plt.show()
```

```python
#filter : FIND_EDGES
edge_im = im.filter(ImageFilter.FIND_EDGES)

plt.imshow(edge_im)
plt.show()
```

```python
#filter : SHARPEN
sharp_im = im.filter(ImageFilter.SHARPEN)

plt.imshow(sharp_im)
plt.show()
```

## merge multiple images

```python
im1 = Image.open("Japan.jpg")
im2 = Image.open("dream house.jfif")
```

```python
im1 = Image.open("Japan.jpg")
im2 = Image.open("dream house.jfif")

fig,axs= plt.subplots(1,2, figsize=(15,8))
axs[0].imshow(im1)
axs[1].imshow(im2)
# plt.imshow(im2)
#plt.imshow(fig)
# plt.show()
plt.show()
```

```python
#checking the size of the images
print(im1.size,im2.size)
```

```python
#changing the size of the image
#resizing  bigger image by giving it the value of smaller image 
im1 = im1.resize((870,580))
im2 = im2.resize((870, 580))
print(im1.size, im2.size)
```

```python
#mentionning colour and portion of the images
merge_img = Image.new("RGB", (2* im1.size[0], im1.size[1]),(250,250,250))
merge_img.paste(im1,(0,0))
merge_img.paste(im2,(im1.size[0],0))
#merge_img.show()

#merging and showing the images
plt.imshow(merge_img)
plt.show()
```
